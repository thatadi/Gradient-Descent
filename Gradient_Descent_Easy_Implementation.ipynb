```python
#consider a regression problem for cgpa & package

X=[6.85,8.34,8.75,9.32,6.5,5.9,7.78]
Y=[4.25,19.20,6.76,11.02,5.2,10.4,4.8]
```

```python
# we know slope of a line is y=mx+c
# m and c are the learnable parameters
# define function to calculate gradients of these parameters
# gradients are derivatives of the loss function w.r.t parameter
# consider loss function to be Mean squared error

def calc_loss(Yhat,Y):
  loss=0
  for i in range(len(Y)):
    loss+=((Yhat[i]-Y[i])**2)
  return loss/len(Y)
def grad_m(X,Y,m,c):
  grad=0
  for i in range(len(X)):
    grad+=((m*X[i]+c)-Y[i])*X[i]
  return grad/len(X)
def grad_c(X,Y,m,c):
  grad=0
  for i in range(len(X)):
    grad=((m*X[i]+c)-Y[i])
  return grad/len(X)
for _ in range(1000):
  m=0
  c=0
  n=0.01
  m-=n*grad_m(X,Y,m,c)
  c-=n*grad_c(X,Y,m,c)
  Yhat=[m*X[i]+c for i in range(len(X))]
  print(calc_loss(Yhat,Y))
```
